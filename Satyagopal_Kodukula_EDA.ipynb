# Which Category of items are mostly sold :

```python
# Merge the datasets on 'ProductId' to associate transactions with categories
merged_df = pd.merge(transactions_df, products_df, on='ProductID')

# Group by Category and count the number of transactions
category_sales = merged_df.groupby('Category')['TransactionID'].count()

# Sort categories by the count of transactions (sold most)
category_sales_sorted = category_sales.sort_values(ascending=False)

# Display the result
print(category_sales_sorted)
```

#Which Region of people are buying the most products :

```python
# Merge transactions with product data to get the product category
merged_df = pd.merge(transactions_df, products_df, on='ProductID')

# Merge the result with the customer data to get the region for each transaction
final_df = pd.merge(merged_df, customers_df, on='CustomerID')

# Group by Region and count the number of transactions
region_sales = final_df.groupby('Region')['TransactionID'].count()

# Sort regions by the number of transactions (buying more items)
region_sales_sorted = region_sales.sort_values(ascending=False)

# Display the result
print(region_sales_sorted)
```

```python
plt.figure(figsize=(8,6))
region_sales_sorted.plot(kind='bar', color='skyblue')
plt.title('Number of Transactions by Region')
plt.xlabel('Region')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.show()
```

# Which Price Range products are sold the most :

```python
# Define price ranges (low, medium, high) based on Price column
bins = [0, 100, 500, 1000, float('inf')]  # You can modify these ranges based on your data
labels = ['Low', 'Medium', 'High', 'Very High']
transactions_df['PriceRange'] = pd.cut(transactions_df['Price'], bins=bins, labels=labels)

# Calculate the total value of each transaction (Quantity * Price)
transactions_df['TotalValue'] = transactions_df['Quantity'] * transactions_df['Price']

# Group by PriceRange and count the number of transactions or sum the total value
price_range_sales = transactions_df.groupby('PriceRange')['TransactionID'].count()

# Sort the result (optional)
price_range_sales_sorted = price_range_sales.sort_values(ascending=False)

# Plotting the result
plt.figure(figsize=(8,6))
price_range_sales_sorted.plot(kind='bar', color='skyblue')
plt.title('Number of Transactions by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=0)
plt.show()
```

# At what date and Time most Sale happed :

```python
# Convert the 'TransactionDate' column to datetime format
transactions_df['TransactionDate'] = pd.to_datetime(transactions_df['TransactionDate'])

# Extract the date and time components (you can change this depending on your need)
transactions_df['Date'] = transactions_df['TransactionDate'].dt.date
transactions_df['Hour'] = transactions_df['TransactionDate'].dt.hour

# Group by Date to get the most sold day
date_sales = transactions_df.groupby('Date')['TransactionID'].count()

# Group by Hour to get the most sold hour
hour_sales = transactions_df.groupby('Hour')['TransactionID'].count()

# Plotting the number of transactions by Hour (most sold hours of the day)
plt.figure(figsize=(10, 6))
hour_sales.plot(kind='bar', color='salmon')
plt.title('Number of Transactions by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=0)
plt.show()
```

# Average Order Value :

```python
# Calculate Average Order Value (AOV) by dividing total revenue by the number of transactions
total_revenue = transactions_df['TotalValue'].sum()
total_transactions = transactions_df['TransactionID'].nunique()
AOV = total_revenue / total_transactions

print(f'Average Order Value (AOV): {AOV}')
```
